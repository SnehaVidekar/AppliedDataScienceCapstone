### PART 1: This notebook is to demonstrate word scraping from website page

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

#Get the website page using BeautifulSoup
res1 = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup1 = BeautifulSoup(res1.content,'lxml')

# Read the postal code table from the website response
table1 = soup1.find_all('table')[0] 
listOfPostalCode = pd.read_html(str(table1))
#print(listOfPostalCode[0].to_json(orient='records'))
#print(listOfPostalCode[0])
```

### Dataframe creation

```python
# Define the dataframe columns
column_names = ['Postalcode','Borough', 'Neighborhood'] 

# Instantiate the dataframe
neighborhoodsDF = pd.DataFrame(columns=column_names)

# Populate the dataframe with postalcode data retrieved from the website
neighborhoodsDF=pd.concat(listOfPostalCode)
neighborhoodsDF.columns= ['Postalcode','Borough', 'Neighborhood']
neighborhoodsDF.head()
```

### Remove rows if Borough column has value 'Not assigned'

```python

# Get names of indexes for which column Borough has value 'Not assigned'
indexNames = neighborhoodsDF[ neighborhoodsDF['Borough'] == 'Not assigned' ].index
# Delete these row indexes from dataFrame
neighborhoodsDF.drop(indexNames , inplace=True)
neighborhoodsDF
```

### To check if Neighborhood column has value 'Not assigned' or 'Not Assigned'

```python
# To check if Neighborhood column has value 'Not assigned' or 'Not Assigned'
mask1= (neighborhoodsDF['Neighborhood'] == "Not assigned")
mask2=(neighborhoodsDF['Neighborhood'] == "Not Assigned")

print("Number of rows in column ""Neighborhood"" has Not assigned value : ",neighborhoodsDF[mask1].shape[0]," or Not Assigned" ,neighborhoodsDF[mask2].shape[0])
```

### Display comma seperated values in Neighborhood column

```python
neighborhoodsDF['Neighborhood'] = neighborhoodsDF['Neighborhood'].str.replace('/',',')
neighborhoodsDF
```

```python
print("Number of rows in the dataframe are", neighborhoodsDF.shape[0], "and columns are" ,neighborhoodsDF.shape[1], ".")
```

### Processing of the venues for clustering purpose

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
# Group venues
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()

print("Processing of venues data is completed.")
```

### Function for sorting venues

```python
import numpy as np
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

print("Top 10 common venues identified.")
neighborhoods_venues_sorted.head()
```

### Clustering Neighborhood

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
#toronto_grouped_clustering
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighbourhood')

toronto_merged.head() 
```

### Cluster visualization

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
